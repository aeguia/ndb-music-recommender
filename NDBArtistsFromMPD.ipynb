### Read NDB Concerts 2016-2021 dataset

```python
data_root = './data/'
ndb_artists_file = os.path.join(data_root, 'artists_ndb.csv')
tracks_file = os.path.join(data_root, 'tracks_ndb.compress')
```

```python
with open(ndb_artists_file, 'rb') as f:
    result = chardet.detect(f.read())

ndb_artists_df = pd.read_csv(ndb_artists_file, encoding=result['encoding'])
```

```python
ndb_artists_df
```

```python
ndb_artists_df['artist_spid'].nunique()
```

```python
columns = ['2016', '2017', '2018', '2019', '2021']
```

```python
ndb_editions = ndb_artists_df.filter(columns).isin([1]).sum()
```

```python
ndb_editions_df= pd.DataFrame({'Edition': ndb_editions.index, 'Total_artists': ndb_editions.values})
```

### Read NDB_MPD Tracks dataset

```python
tracks_df = pd.read_pickle(tracks_file, compression='gzip')
```

```python
tracks_df
```

```python
tracks_df['artist_spid'] = [ReadJSONFiles.parseURI(uri) for x, uri in enumerate(tracks_df['artist_uri'])]
```

#### Filter NDB_MPD Tracks dataset with NDB Artists

```python
ndb_mpd_tracks_df = tracks_df[tracks_df['artist_spid'].isin(ndb_artists_df['artist_spid'].to_list())]
```

```python
ndb_mpd_tracks_df
```

```python
# Number of Tracks per NDB Artist in NDB_MPD dataset

ndb_mpd_tracks_df.groupby('artist_name')['tid'].count().sort_values()
```

```python
def getArtistName(artist_spid, df):
    """
    Return Artist Name
    :param uri: artist Spotify ID
    :param df: dataframe
    """     
    return df.loc[df['artist_spid']==artist_spid]['artist_name'].iloc[0]
```

#### Total number of unique artist playing in NDB Festival

```python
ndb_artists_spid_list = ndb_artists_df['artist_spid'].to_list()
```

```python
len(ndb_artists_spid_list)
```

```python
# List of Artist names playing at NDB

[getArtistName(spid, ndb_artists_df) for spid in ndb_artists_spid_list]
```

#### Total number of unique artist from NDB in NDB_MPD dataset

```python
ndb_artists_in_mpd_list = ndb_mpd_tracks_df['artist_spid'].unique()
```

```python
len(ndb_artists_in_mpd_list)
```

```python
# List of Artist names playing at NDB in NDB_MPD dataset

[getArtistName(spid, ndb_mpd_tracks_df) for spid in ndb_artists_in_mpd_list]
```

#### Number of  NDB artist NOT in NDB_MPD dataset

```python
ndb_artists_not_mpd_list = list(set(ndb_artists_spid_list) - set(ndb_artists_in_mpd_list))
```

```python
len(ndb_artists_not_mpd_list)
```

```python
# List of Artist names playing at NDB not in MPD dataset

[getArtistName(spid, ndb_artists_df) for spid in ndb_artists_not_mpd_list]
```

#### Merge NDB_MPD tracks from NDB Artists with NDB Editions Features

```python
merge_columns = ['artist_spid', 'artist_followers', 'artist_popularity', 'artist_genres', '2016', '2017', '2018', '2019', '2021']
```

```python
ndb_mpd_tracks_df = ndb_mpd_tracks_df.merge(ndb_artists_df[merge_columns], how='left', on=['artist_spid'])
```

```python
ndb_mpd_tracks_df
```

```python
print(f'Export DF {len(ndb_mpd_tracks_df)} NDB Artists Tracks in MPD to Pickle')
ndb_mpd_tracks_df.to_pickle('./data/ndb_mpd_tracks.compress', compression='gzip')
```
