```python
import os
import datetime
import pandas as pd
import numpy as np

pd.set_option("max_rows", 10)
data_root = "../data/"
```

```python
concerts = os.path.join(data_root, "artists_ndb.csv")
df_concerts_ndb = pd.read_csv(concerts)
```

```python
tracks = os.path.join(data_root, "artist_tracks.csv")
df_tracks_ndb = pd.read_csv(tracks)
```

```python
year = []

for i in range(0, len(df_tracks_ndb['album_release_date'])):
    year.append(df_tracks_ndb['album_release_date'][i][0:4])
```

```python
df_tracks_ndb['year'] = year
```

```python
df_all_tracks = df_tracks_ndb.merge(df_concerts_ndb, how='left', on=['artist_spid','artist_name'])
```

```python
# Handle missing values derived from 'artist_spid' --> 'artist_popularity' | 'artist_followers' | 'artist_genres'

df_all_tracks[df_all_tracks['artist_spid'].isna()]
```

```python
# Get album popularity infered by mean from track popularity

lalaland_popularity = df_all_tracks[df_all_tracks['artist_spid'].isna()]['track_popularity'].mean()
lalaland_popularity
```

```python
# Fill artist_popularity NaN with album_popularity

df_all_tracks.loc[df_all_tracks[df_all_tracks['artist_spid'].isna()].index,'artist_popularity'] = lalaland_popularity
```

```python
# Fill artist_followers NaN with 0

df_all_tracks.loc[df_all_tracks[df_all_tracks['artist_spid'].isna()].index,'artist_followers'] = 0
```

```python
# Fill artist_genres NaN with soundtrack

genre = []
genre.append(['soundtrack'])
df_all_tracks.loc[df_all_tracks[df_all_tracks['artist_spid'].isna()].index,'artist_genres'] = genre
```

```python
df_all_tracks['artist_popularity'] = df_all_tracks['artist_popularity'].astype(int)
df_all_tracks['artist_followers'] = df_all_tracks['artist_followers'].astype(int)
```

```python
df_all_tracks.groupby('artist_name')[['track_popularity','artist_popularity','artist_followers']].describe()
```
